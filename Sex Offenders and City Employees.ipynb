```python
import pandas as pd
```

```python
offenders = pd.read_csv('Sex_Offenders.csv')
```

```python
employees = pd.read_csv('Current_Employee_Names__Salaries__and_Position_Titles.csv')
```

```python
offenders
```

```python
employees
```

```python
#Find the intersection of the two datasets
overlap = None

for offender in offenders.iterrows():
    first = offender[1]['FIRST']
    last = offender[1]['LAST'] 
    
    #TODO: Make sure we're not matching the first name on the last name, and vice versa
    extracts = employees.loc[
        employees['Name'].str.contains(first) &
        employees['Name'].str.contains(last)
    ]
    if len(extracts) > 0:
        if overlap is None:
            overlap = extracts
        else:
            overlap = pd.concat([overlap, extracts])
            
#Deduplicate
overlap = overlap[~overlap.index.duplicated(keep='first')]
```
