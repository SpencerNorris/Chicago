```python
#Set address fields to lower-case
vacant_buildings['Property Address'] = vacant_buildings['Property Address'].str.lower()
business_licenses['ADDRESS'] = business_licenses['ADDRESS'].str.lower()

#Narrow down vacancies to...
narrowed_vacancies = vacant_buildings[
    vacant_buildings['Issued Date'] >= datetime.date(2017,1,1) #year=2017
] 
```

```python
#Merge the 2018 vacancies with the active licenses based on whether
# the vacancy address is a substring of the business license address
possible_vacant_businesses = None

for vacancy in narrowed_vacancies.iterrows():
    #Pull list of businesses that have the vacancy address as a substring
    extracted_businesses = business_licenses.loc[
        business_licenses['ADDRESS'].str.contains(vacancy[1]['Property Address'])
    ]
    
    #Check if we found any
    if len(extracted_businesses) > 0:
        #Initial population
        if possible_vacant_businesses is None:
            possible_vacant_businesses = extracted_businesses
        #Appending to existing dataframe
        else:
            possible_vacant_businesses = pd.concat(
                [possible_vacant_businesses, extracted_businesses],
                ignore_index=True
            )
```
